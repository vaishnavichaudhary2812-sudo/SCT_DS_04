```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('traffic_accidents.csv')

print(df.shape)
print(df.isnull().sum())

df = df.dropna()

plt.figure(figsize=(8,5))
sns.countplot(x='weather_condition', data=df, order=df['weather_condition'].value_counts().head(10).index)
plt.xticks(rotation=45)
plt.title('Accidents by Weather Condition')
plt.show()

plt.figure(figsize=(8,5))
sns.countplot(x='roadway_surface_cond', data=df, order=df['roadway_surface_cond'].value_counts().head(10).index)
plt.xticks(rotation=45)
plt.title('Road Surface Condition')
plt.show()

plt.figure(figsize=(10,5))
sns.countplot(x='crash_hour', data=df)
plt.title('Accidents by Hour of Day')
plt.show()

plt.figure(figsize=(8,5))
sns.countplot(x='crash_day_of_week', data=df)
plt.title('Accidents by Day of Week')
plt.show()

plt.figure(figsize=(8,5))
sns.countplot(x='lighting_condition', data=df, order=df['lighting_condition'].value_counts().index)
plt.xticks(rotation=45)
plt.title('Lighting Condition')
plt.show()

plt.figure(figsize=(10,5))
sns.countplot(x='prim_contributory_cause',
              data=df,
              order=df['prim_contributory_cause'].value_counts().head(10).index)
plt.xticks(rotation=90)
plt.title('Top Contributing Factors')
plt.show()

plt.figure(figsize=(8,5))
sns.countplot(x='most_severe_injury',
              data=df,
              order=df['most_severe_injury'].value_counts().index)
plt.xticks(rotation=45)
plt.title('Most Severe Injury')
plt.show()

plt.figure(figsize=(10,6))
sns.heatmap(df[['injuries_total',
                'injuries_fatal',
                'injuries_incapacitating',
                'injuries_non_incapacitating',
                'crash_hour']].corr(),
            annot=True,
            cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
